```python
%matplotlib inline
import sys
print(sys.version)
import numpy as np
print(np.__version__)
import pandas as pd
print(pd.__version__)
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv('NYPD_Complaint_Data_Historic.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['PARKS_NM','HADEVELOPT'],axis =1)
```

```python
df.columns
```

```python
# Data sanitization
# column CMPLNT_FR_DT:
# Exact date of occurrence for the reported event (or starting date of occurrence, if CMPLNT_TO_DT exists)

df.CMPLNT_FR_DT.value_counts().sort_index()
```

```python
df.CMPLNT_FR_DT.isnull().any()
```

```python
df=df.dropna(subset = ['CMPLNT_FR_DT'])
df.shape
```

```python
df['CMPLNT_FR_DT'] = pd.to_datetime(df['CMPLNT_FR_DT'],errors = 'coerce')
```

```python
df.dtypes
```

```python
df = df[(df['CMPLNT_FR_DT'] >= '2006-1-1') & (df['CMPLNT_FR_DT'] <= '2016-12-31')]
df = df.dropna(subset = ['CMPLNT_FR_DT'])
df.shape
```

```python
# OFNS_DESC:Description of offense corresponding with key code

df.OFNS_DESC.isnull().any()
```

```python
df=df.dropna(subset = ['OFNS_DESC'])
df.shape
```

```python
crime = df.OFNS_DESC.value_counts().head(10)
crime
```

```python
crime.plot(kind='bar')
plt.ylabel("Offense Counts")
```

```python
# OFNS_DESC by CMPLNT_FR_DT year

crime_vs_year = df.groupby(df.CMPLNT_FR_DT.dt.to_period("Y")).OFNS_DESC.count()
print(crime_vs_year)
```

```python
crime_vs_year.plot()
plt.ylabel("Offense Counts")
plt.xlabel("Offense Occurence Year")
```

```python
# RPT_DT:Date event was reported to police 

df.RPT_DT.isnull().any()
```

```python
df['RPT_DT'] = pd.to_datetime(df['RPT_DT'],errors = 'coerce')
```

```python
df.RPT_DT.value_counts().sort_index()
```

```python
print(df.RPT_DT.min(), df.RPT_DT.max())
```

```python
# OFNS_DESC by RPT_DT year

offense_vs_RPT_year = df.groupby(df.CMPLNT_FR_DT.dt.to_period("Y")).OFNS_DESC.count()
print(offense_vs_RPT_year)
```

```python
offense_vs_RPT_year.plot()
plt.ylabel("Offense Counts")
plt.xlabel("Offense Report Year")
```

```python
# by month
offense_by_month = df.groupby(df.RPT_DT.dt.to_period("M")).OFNS_DESC.count()
offense_by_month
```

```python
# average number of daily records per month

daily = df.groupby(df.RPT_DT.dt.to_period("D")).OFNS_DESC.count()
```

```python
daily
```

```python
# monthly average
monthly = daily.groupby(pd.Grouper(freq='M')).mean()
```

```python
monthly['2007'].plot()
```

```python
ax = pd.DataFrame(monthly)
ax
```

```python
dm = pd.DataFrame(offense_by_month)
pm = pd.pivot_table(dm, index=dm.index.month, columns=dm.index.year, aggfunc='sum')
pm
```

```python
pm.plot(figsize=(20,10))
plt.xlabel("Month")
plt.ylabel("Crime Counts")
```

```python
ad = dm.reset_index()
ad.shape
ad.columns

ad['monthly_average'] = ad.OFNS_DESC/ad.RPT_DT.dt.daysinmonth
ad = ad.drop(['OFNS_DESC'],axis =1)
ad
```

```python
ad.set_index('RPT_DT', inplace=True)
ad
```

```python
df2 = pd.pivot_table(ad, index=ad.index.month, columns=ad.index.year, aggfunc='sum')
df2
```

```python
df2['monthly_average_mean']=df2.mean(axis=1)
df2['average_low_temperature'] = [27,28,35,45,54,64,69,68,61,50,42,32]
df2
```

```python
df2.monthly_average_mean.plot(kind='bar',figsize=(20,10),ylim=(1200, 1500))
plt.xlabel("Month")
plt.ylabel("Offense Monthly Mean over 2006-2016")
```

```python
df2['monthly_average_mean'].corr(df2['average_low_temperature'])
```

```python
df2.plot(figsize=(20,10))
plt.xlabel("Month")
plt.ylabel("Offense Monthly Average")
```
